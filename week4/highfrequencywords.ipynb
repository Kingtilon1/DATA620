### Choose a corpus of interest: I will choose the names corpus

```python
import nltk
from nltk.corpus import names
import matplotlib.pyplot as plt
from collections import Counter

# Combine the names from both male and female datasets
all_names = names.words('male.txt') + names.words('female.txt')
```

### Counting the unique words

```python
unique_words = set(all_names)
total_unique_words = len(unique_words)
print(f'Total unique words: {total_unique_words}')
```

### Most common words

```python
word_counts = Counter(all_names)
total_words = sum(word_counts.values())

half_total_words = total_words / 2
current_count = 0
unique_word_count_for_half = 0

for word, count in word_counts.most_common():
    current_count += count
    unique_word_count_for_half += 1
    if current_count >= half_total_words:
        break

print(f'Number of unique words that represent half of total words: {unique_word_count_for_half}')
```

### Identify the 200 highest frequency words

```python
top_words = word_counts.most_common(200)
print(f'Top 200 words: {top_words}')
```

### Create graph of relative frequency

```python
words, counts = zip(*top_words)
relative_frequencies = [count / total_words for count in counts]

plt.figure(figsize=(15, 8))
plt.bar(words, relative_frequencies)
plt.xlabel('Words')
plt.ylabel('Relative Frequency')
plt.title('Relative Frequency of the Top 200 Words')
plt.xticks(rotation=90)
plt.show()
```

### 6: Does the observed relative frequency of these words follow Zipfs law? Explain.
We looked at how often names appear in our dataset and compared it to Zipf's law. Zipf's law suggests that the most common word appears much more frequently than less common words, following an inverse relationship.

When we plotted the top 200 names and their frequencies, we saw that the trend matches what Zipfs law predicts. This means a few names are very common, and many names are rare. This pattern shows that the names dataset does follow Zipfs law. For example, the most common names like 'James' and 'Mary' appear much more frequently than less common names. The graph visually supports this by showing a steep drop-off from the most frequent names to the less frequent ones.
